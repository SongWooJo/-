```python
from sklearn.datasets import load_iris
import pandas as pd
import  numpy as np

#데이터셋 로드
iris = load_iris()
df = pd.DataFrame(data=np.c_[iris['data'],iris['target']],columns=['sepal length','sepal width','petal length','petal width','class'])
df
```

```python
from sklearn.model_selection import KFold

X = np.array(df.iloc[:, :-1])  #class 열 제외한 feature 열들 모음 -> array 변환
y = df['class']

# split 개수, 셔플 여부 및 seed 설정
kf = KFold(n_splits=5, shuffle=True, random_state=50)

# split 개수 스텝 만큼 train, test 데이터셋을 매번 변환
for train_index, test_index in kf.split(X):
    print(train_index)
    #print('\n',  len(train_index))
    print(test_index)
    #print('\n', len(test_index))
```

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score

accuracy_history = []

# K-Fold 검증 과정으로 실제 랜덤 포레스트 모델을 학습하여 정확도 평균을 내는 방법
for train_index, test_index in kf.split(X):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model = RandomForestClassifier(n_estimators=5, random_state=0) # 모델 선언
    model.fit(X_train, y_train) # 모델 학습

    y_pred = model.predict(X_test)  # 모델 예측값

    accuracy_history.append(accuracy_score(y_test, y_pred)) # 정확도 측정 및 기록

# 출력 결과
print("각 분할의 정확도 기록 : ", accuracy_history)
print("평균 정확도 : ", np.mean(accuracy_history))
```

```python
from sklearn.model_selection import StratifiedKFold

accuracy_history = []

# split 개수, 셔플 여부 및 seed 설정
str_kf = StratifiedKFold(n_splits=5, shuffle=True, random_state=50)

# split 개수 스텝 만큼 train, test  데이터셋을 매번 분할
for train_index, test_index in str_kf.split(X, y):
    X_train, X_test = X[train_index], X[test_index]
    y_train, y_test = y[train_index], y[test_index]

    model = RandomForestClassifier(n_estimators=5, random_state=0) # 모델 선언
    model.fit(X_train, y_train) # 모델 학습

    y_pred = model.predict(X_test)  # 모델 예측값

    accuracy_history.append(accuracy_score(y_test, y_pred)) # 정확도 측정 및 기록

# 출력 결과
print("각 분할의 정확도 기록 : ", accuracy_history)
print("평균 정확도 : ", np.mean(accuracy_history))
```
